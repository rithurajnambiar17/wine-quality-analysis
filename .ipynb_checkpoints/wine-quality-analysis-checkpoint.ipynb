```python
data.drop(columns = ['sulfur.dioxide'], inplace = True)
```

```python
data
```

```python
data.reset_index(inplace = True)
data.drop(columns = ['index'], inplace = True)
data
```

```python
for i in range(0, 6497):
    if data['quality'][i] <= 6:
        data['quality'][i] = 'bad'
    elif data['quality'][i] >= 7:
        data['quality'][i] = 'good'
data
```

```python
data.quality.value_counts()
```

```python
dummy = pd.get_dummies(data.quality)
data['bad'] = dummy['bad']
data['good'] = dummy['good']
```

```python
data.drop(columns = ['quality'], inplace = True)
```

```python
data
```

```python
X = data.drop(columns = ['bad', 'good'])
y = data['']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42, shuffle = True)
print(f'''
Shape of X_train : {X_train.shape}
Shape of X_test : {X_test.shape}
Shape of y_train : {y_train.shape}
Shape of y_test : {y_test.shape}
''')
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
#creating model
model = Sequential()
model.add(Dense(12, input_dim = 12,activation="relu"))
model.add(Dense(8, activation="relu"))
model.add(Dense(1, activation="sigmoid"))
```

```python
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=10, batch_size=10)
```

```python
scores = model.evaluate(X_train, y_train)
print(f"{model.metrics_names[1], scores[1]*100}")
```

```python
predict = model.predict(X_test)
```

```python
metric = tf.keras.metrics.Accuracy()
metric.update_state(y_test , predict)
metric.result().numpy()
```
